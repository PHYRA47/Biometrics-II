```python
import cv2
import numpy as np
import tensorflow as tf
from pycocotools.coco import COCO
from pycocotools import mask as maskUtils
from sklearn.metrics import average_precision_score

import matplotlib.pyplot as plt
```

```python

# Path to the annotation file
annFile = '../data/person_keypoints_val2017.json'

# Initialize COCO API
coco = COCO(annFile)
```

```python
# Get all image IDs
imgIds = coco.getImgIds()

# Load images
images = coco.loadImgs(imgIds)
```

```python

for img in images[:2]:
    # Load image
    image = cv2.imread('../data/val2017/' + img['file_name'])
    plt.imshow(image)
    plt.axis('off')

    # Get annotation IDs for the image
    annIds = coco.getAnnIds(imgIds=img['id'], iscrowd=False); print(annIds)
    anns = coco.loadAnns(annIds); print(anns)

    # Display keypoints
    coco.showAnns(anns)
    plt.show()
```
